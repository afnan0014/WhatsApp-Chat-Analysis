```python
import re
import pandas as pd
```

```python
f=open('WhatsApp Chat with Batch H.txt','r',encoding='utf-8')
```

```python
data=f.read()
```

```python
print(data)
```

```python
pattern = r"\d{1,2}/\d{1,2}/\d{2}, \d{1,2}:\d{2}[\u202f\s]?[APMapm]{2}"
```

```python
messages=re.split(pattern, data)[1:]
messages
```

```python
dates=re.findall(pattern, data)
# Clean each date string by removing \u202f
cleaned_dates = [d.replace('\u202f', '') for d in dates]

print(cleaned_dates)
```

```python
print(len(cleaned_dates))
print(len(messages))
```

```python
df = pd.DataFrame({'user_message': messages, 'message_date': cleaned_dates})
    # convert message_date type
df['message_date'] = pd.to_datetime(df['message_date'], format="%m/%d/%y, %I:%M%p")
df['message_date'] = df['message_date'].dt.strftime("%m/%d/%Y, %I:%M %p")

df.rename(columns={'message_date': 'date'}, inplace=True)
df.head()
```

```python
df.shape
```

```python
users = []
messages = []
for message in df['user_message']:
        entry = re.split('([\w\W]+?):\s', message)
        if entry[1:]:  # user name
            users.append(entry[1])
            messages.append(" ".join(entry[2:]))
        else:
            users.append('group_notification')
            messages.append(entry[0])
```

```python
df['user'] = users
df['message'] = messages
df.drop(columns=['user_message'], inplace=True)
df
```

```python
# First, ensure that 'date' is in string format if it's not
df['date'] = df['date'].astype(str)

# Convert the 'date' column to datetime using the specified format for extraction
df['date'] = pd.to_datetime(df['date'], format='%m/%d/%Y, %I:%M %p', errors='coerce')

# Extract components from the datetime in 12-hour format
df['only_date'] = df['date'].dt.date
df['year'] = df['date'].dt.year
df['month_num'] = df['date'].dt.month
df['month'] = df['date'].dt.month_name()
df['day'] = df['date'].dt.day
df['day_name'] = df['date'].dt.day_name()

# Extract hour, minute in 12-hour format and AM/PM
df['hour'] = df['date'].dt.strftime('%I')  # Extract hour in 12-hour format (01-12)
df['minute'] = df['date'].dt.strftime('%M')  # Extract minute
df['AM_PM'] = df['date'].dt.strftime('%p')  # Extract AM/PM

# Optionally, convert the date back to the original format if you want to keep the string format
df['date'] = df['date'].dt.strftime('%m/%d/%Y, %I:%M %p')



df
```

```python
period = []
for index, row in df.iterrows():
    hour = int(row['hour'])  # Ensure 'hour' is treated as an integer
    
    # Convert hour to 12-hour format and append AM/PM
    hour_in_12hr = hour % 12  # Converts hour to 12-hour format (0-11)
    am_pm = 'AM' if hour < 12 else 'PM'
    
    # Handle special cases for midnight and noon
    if hour == 0:  # Midnight case (12:00 AM)
        period.append(str('12') + " AM - " + str('01') + " AM")
    elif hour == 12:  # Noon case (12:00 PM)
        period.append(str('12') + " PM - " + str('01') + " PM")
    else:
        next_hour_in_12hr = (hour + 1) % 12
        next_am_pm = 'AM' if (hour + 1) < 12 else 'PM'
        period.append(f"{hour_in_12hr} {am_pm} - {next_hour_in_12hr} {next_am_pm}")

df['period'] = period
df

```
